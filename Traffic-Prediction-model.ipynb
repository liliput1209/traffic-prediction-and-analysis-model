```python
# Install necessary libraries
!pip install optuna
!pip install xgboost

# Imports
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import optuna
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from xgboost import XGBRegressor

# Upload file
uploaded = files.upload()

# Load dataset
data = pd.read_csv('Traffic Final Data.csv')

# Check columns
print(data.columns)

#  If 'Date' is available, parse it
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'], errors='coerce')  # Safe conversion
    data['Month'] = data['Date'].dt.month
    data['DayOfWeek'] = data['Date'].dt.dayofweek
else:
    raise ValueError("Date column is missing. Can't create Month and DayOfWeek features.")

# Split by Location
asroga_data = data[data['Location'] == 'Asroga Toll Plaza']
bara_data = data[data['Location'] == 'Bara Toll Plaza']



# Optimization + Training Function
def bayesian_optimize_and_compare(data_subset, title):
    #  New feature set
    X = data_subset[['Total Vehicles', 'Month', 'DayOfWeek']]
    Y = data_subset[[
        'Car, Jeep, Van / Light Motor Vehicle',
        'Light Commercial Vehicle, Lights Goods Vehicle or Mini Bus',
        'Bus or Truck (2 Axle)',
        'Bus or Truck (3 Axle)',
        'Heavy Construction Machinery (HCM) or Earth Moving Equipment (EME) or Multi Axle Vehicle (MAV) (4 to 6 axles)',
        'Over Size Vehicles (seven or more axles)'
    ]]

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    #  Random Forest Objective
    def rf_objective(trial):
        model = RandomForestRegressor(
            n_estimators=trial.suggest_int('n_estimators', 100, 300),
            max_depth=trial.suggest_int('max_depth', 5, 20),
            min_samples_split=trial.suggest_int('min_samples_split', 2, 10),
            random_state=42
        )
        multi_model = MultiOutputRegressor(model)
        multi_model.fit(X_train, y_train)
        preds = multi_model.predict(X_test)
        return r2_score(y_test, preds)

    print(f"\n Optimizing RandomForest for {title}...")
    rf_study = optuna.create_study(direction='maximize')
    rf_study.optimize(rf_objective, n_trials=20)

    print(f" Best RandomForest Params: {rf_study.best_trial.params}")

    #  XGBoost Objective
    def xgb_objective(trial):
        model = XGBRegressor(
            n_estimators=trial.suggest_int('n_estimators', 100, 300),
            max_depth=trial.suggest_int('max_depth', 3, 10),
            learning_rate=trial.suggest_float('learning_rate', 0.01, 0.3),
            objective='reg:squarederror',
            random_state=42
        )
        multi_model = MultiOutputRegressor(model)
        multi_model.fit(X_train, y_train)
        preds = multi_model.predict(X_test)
        return r2_score(y_test, preds)

    print(f"\n Optimizing XGBoost for {title}...")
    xgb_study = optuna.create_study(direction='maximize')
    xgb_study.optimize(xgb_objective, n_trials=20)

    print(f" Best XGBoost Params: {xgb_study.best_trial.params}")

    # Train best models
    best_rf = RandomForestRegressor(**rf_study.best_trial.params, random_state=42)
    multi_rf = MultiOutputRegressor(best_rf)
    multi_rf.fit(X_train, y_train)
    y_pred_rf = multi_rf.predict(X_test)

    best_xgb = XGBRegressor(**xgb_study.best_trial.params, objective='reg:squarederror', random_state=42)
    multi_xgb = MultiOutputRegressor(best_xgb)
    multi_xgb.fit(X_train, y_train)
    y_pred_xgb = multi_xgb.predict(X_test)

    #  Evaluation
    mae_rf = mean_absolute_error(y_test, y_pred_rf)
    r2_rf = r2_score(y_test, y_pred_rf)

    mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
    r2_xgb = r2_score(y_test, y_pred_xgb)

    print(f"\n RandomForest Results for {title}: MAE: {mae_rf:.2f}, R2: {r2_rf:.4f}")
    print(f"XGBoost Results for {title}: MAE: {mae_xgb:.2f}, R2: {r2_xgb:.4f}")

    print("\n Final Winner:")
    if r2_rf > r2_xgb:
        print(f" Random Forest wins with R2 = {r2_rf:.4f}!")
        best_pred = y_pred_rf
    else:
        print(f" XGBoost wins with R2 = {r2_xgb:.4f}!")
        best_pred = y_pred_xgb

    return best_pred, Y.columns

#  Plotting function
def plot_vehicle_type_graphs(data_subset, title, best_pred):
    vehicle_types = [
        'Car, Jeep, Van / Light Motor Vehicle',
        'Light Commercial Vehicle, Lights Goods Vehicle or Mini Bus',
        'Bus or Truck (2 Axle)',
        'Bus or Truck (3 Axle)',
        'Heavy Construction Machinery (HCM) or Earth Moving Equipment (EME) or Multi Axle Vehicle (MAV) (4 to 6 axles)',
        'Over Size Vehicles (seven or more axles)'
    ]

    fig, axes = plt.subplots(3, 2, figsize=(15, 15))
    axes = axes.flatten()

    for i, vehicle in enumerate(vehicle_types):
        ax = axes[i]
        ax.plot(np.arange(len(best_pred)), data_subset[vehicle].values[:len(best_pred)], label='Actual', linestyle='-')
        ax.plot(np.arange(len(best_pred)), best_pred[:, i], label='Predicted', linestyle='--')
        ax.set_title(f'{vehicle} - {title}')
        ax.set_xlabel('Samples')
        ax.set_ylabel('Vehicle Count')
        ax.legend()
        ax.grid(True)

    plt.tight_layout()
    plt.show()

# Run for Asroga
print("\n Running for Asroga Toll Plaza...")
best_pred_asroga, vehicle_types_asroga = bayesian_optimize_and_compare(asroga_data, "Asroga")
plot_vehicle_type_graphs(asroga_data, "Asroga", best_pred_asroga)

# Run for Bara
print("\n Running for Bara Toll Plaza...")
best_pred_bara, vehicle_types_bara = bayesian_optimize_and_compare(bara_data, "Bara")
plot_vehicle_type_graphs(bara_data, "Bara", best_pred_bara)
```
